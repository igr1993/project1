```python
# Dependencies and Setup
import pandas as pd
from functools import reduce

# File to load
file1 = "Data/regional-us-weekly-2020-11-27--2020-12-04.csv"
file2 = "Data/regional-us-weekly-2020-12-04--2020-12-11.csv"
file3 = "Data/regional-us-weekly-2020-12-11--2020-12-18.csv"
file4 = "Data/regional-us-weekly-2020-12-18--2020-12-25.csv"
file5 = "Data/regional-us-weekly-2020-12-25--2021-01-01.csv"

# Read files and store into data frame
song_data1 = pd.read_csv(file1)
song_data2 = pd.read_csv(file2)
song_data3 = pd.read_csv(file3)
song_data4 = pd.read_csv(file4)
song_data5 = pd.read_csv(file5)

# Drop "Position" column for each data frame
df1 = song_data1.drop(['Position'], axis=1)
df2 = song_data2.drop(['Position'], axis=1)
df3 = song_data3.drop(['Position'], axis=1)
df4 = song_data4.drop(['Position'], axis=1)
df5 = song_data5.drop(['Position'], axis=1)

data_frames = [df1, df2, df3, df4, df5]

# Combine the data into a single dataset
df_merged = reduce(lambda left,right: pd.merge(left,right,on=['URL'],
                                            how='outer'), data_frames)
df_merged.head()
```

```python
total_streams = df_merged.sum(axis = 1, skipna = True)
total_streams.head()
```

```python
df_merged['Total Streams'] = total_streams
df_merged.head()
```

```python
df_clean = df_merged.drop(['Track Name_x', 'Artist_x', 'Streams_x', 'URL', 
                           'Track Name_y', 'Artist_y', 'Streams_y', 'Streams'], axis=1)
df_clean.head()
```

```python
df_final = df_clean.sort_values('Total Streams', ascending=False)
df_final.head(10)
```

```python
df_final.to_csv(r'Export/Christmas2020.csv', index = False)
```
